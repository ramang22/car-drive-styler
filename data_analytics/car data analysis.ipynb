```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df =  pd.read_csv("./data/test123.csv",delimiter=";") 
df['date'] =  pd.to_datetime(df['date'], format='%m/%d/%Y %H:%M:%S')
df.index = df['date']
df = df.drop(columns={"date"})
```

```python
df.head()
```

```python
df.info()
```

```python
for name in df.columns:
    df[name].plot(linewidth=1.5, title=name.upper());
    plt.show()
```

```python
ax = sns.heatmap(
    df.corr(), 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(100, 200, n=200),
    square=True,
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    horizontalalignment='right'
);
```

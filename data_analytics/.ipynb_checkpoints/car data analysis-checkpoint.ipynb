```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df =  pd.read_csv("./data/AUP2.csv",delimiter=";") 
df['date'] =  pd.to_datetime(df['date'], format='%m/%d/%Y %H:%M:%S')
df.index = df['date']
df = df.drop(columns={"date"})
for name in df.columns:
    df[name].plot(linewidth=1.5, title=name.upper());
    plt.show()
for name in df.columns:
    if name == "speed" or name == "gforce":
        continue
    sns.scatterplot(data=df, x="speed", y=name, hue="gforce")
    plt.title("Speed x "+str(name))
    plt.show()
```
